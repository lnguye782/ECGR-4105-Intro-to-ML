```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_url = 'https://raw.githubusercontent.com/lnguye782/ECGR-4105-Intro-to-ML/refs/heads/main/HW3/cancer.csv'
data = pd.read_csv(file_url)

data.head()
```

```python
# Separate features and target variable (30 input features / 1 output target)
X = data.drop(columns=['target'])
Y = data['target']
```

```python
# Split the data set into Training Data (80%) and Test Data (20%)
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Scale the data between 0 and 1 to get better accuracy
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# Fit a Naive Bayes model to the data
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
# Predict on the test data
Y_pred = model.predict(X_test)
```

```python
# Evaluate the model with L2 regularization using model evaluation metrics: accuracy, precision, recall, and F1 score
from sklearn import metrics

print("Accuracy with Naive Bayes:",metrics.accuracy_score(Y_test, Y_pred))
print("Precision with Naive Bayes:",metrics.precision_score(Y_test, Y_pred))
print("Recall with Naive Bayes:",metrics.recall_score(Y_test, Y_pred))
print("F1 Score with Naive Bayes:",metrics.f1_score(Y_test, Y_pred))
```
